```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
from PIL import Image, ImageDraw
import cv2
import numpy as np

with open('bbox_stack.pkl', 'rb') as f:
    bbox = pickle.load(f)
with open('layout_stack.pkl', 'rb') as f:
    layout = pickle.load(f)
print(bbox)
print(layout)
# 
image_path = "/root/workspace/donut/synthdog/outputs/SynthDoG_zh/train/image_0.jpg"
image = Image.open(image_path)

# bbox
draw = ImageDraw.Draw(image)
draw.rectangle(bbox, outline="red")

#layout
for (x, y, w, h), _ in layout:
    draw.rectangle([x, y, x+w, y+h], outline="blue")
# Pillownumpy
image = np.array(image)

# 
plt.imshow(image)
plt.show()
```
